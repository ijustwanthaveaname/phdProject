```python
wb = xlrd.open_workbook("华农评分项目.xls", formatting_info=True)
# identify colored cell 
sheet = wb.sheets()[0]
bgcol = np.zeros([sheet.nrows, sheet.ncols])
for row in range(sheet.nrows):
    for column in range(sheet.ncols):
        cell = sheet.cell(row, column)
        fmt = wb.xf_list[cell.xf_index]
        bgcol[row, column] = fmt.background.pattern_colour_index
colormask = pd.DataFrame(bgcol)
print(np.unique(colormask))
```

```python
colored_cell = np.where(colormask != 64)
```

```python
df = pd.read_excel("./华农评分项目.xls", header=None)
df.head()
df[0] = df[0].str.replace(" ", "")
df.head()
```

```python
# df[0].str.split("♀", expand=True)
cage = df[0].str.slice(0, 1).fillna(method='ffill', axis=0) 
dietary = df[0].str.slice(1, 3).fillna(method='ffill', axis=0) 
sex = df[0].str.slice(3, 4).fillna(method='ffill', axis=0) 
group = df[0].str.slice(4).fillna(method='ffill', axis=0) 
```

```python
df_fill = df.fillna(method='ffill', axis=0)  # bfill ,backward or upward
df_fill
```

```python
df_temp = df_fill.drop_duplicates(subset=[1], keep='last').iloc[:, [1, 31, 32]]
df_temp.columns = ["id", "体温", "体重"]
df_temp.head()
```

```python
df_combine = pd.concat([df_fill.drop(0, axis=1), cage, dietary, sex], ignore_index=True, axis=1)
df_new = df_combine[~df_combine[1].isin([0, 0.5, 1])]
df_new.index = range(df_new.shape[0])
df_new.iloc[:, 1:-3] = 0
df_new
```

```python
for x, y in zip(*colored_cell):
    df_new.iloc[df_new[0] == df_combine.iloc[x, 0], y-1] = df_fill.iloc[x, y]
```
